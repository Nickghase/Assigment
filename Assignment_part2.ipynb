# Matching Problem

```python
import pandas as pd # for dataframe
```

```python
amaz= pd.read_csv('amz_com-ecommerce_sample.csv',encoding='latin-1')  #read amazon data
amaz.head()
```

```python
flip = pd.read_csv('flipkart_com-ecommerce_sample.csv',encoding='latin-1') #read flipkart data
flip.head()
```

# Extrcting useful columns

```python
amazon_data = amaz[['product_name','retail_price','discounted_price']]
flipkart_data = flip[['product_name','retail_price','discounted_price']]
```

## Sentence Transformer : It is used to create Text embeddings,These embeddings is compared with cosine-similarity to find product names with a similar meaning

```python
!pip install sentence_transformers
```

```python
from sentence_transformers import SentenceTransformer, util
import torch
```

```python
# prebuild model is used for text embedding
model = SentenceTransformer('all-MiniLM-L6-v2') 

# Compute embedding for both Amazon and Flipkart product names
amazon_prod_embs = model.encode(amazon_data.product_name,convert_to_tensor=True)
flipkart_prod_embs = model.encode(flipkart_data.product_name,convert_to_tensor=True)

# cosine similarities between amazon and flipkart
cosine_sim_scores = util.cos_sim(amazon_prod_embs,flipkart_prod_embs)
```

```python
cosine_sim_scores
```

```python
# Storing the index of a similar Flipkart product corresponding to each Amazon product 
similar_prod_indexes = []

for i in range(20000): # 20000 because we have 20000 rows
    max_score_idxs = torch.topk(cosine_sim_scores[i],k=1,largest=True,sorted=True).indices
    similar_prod_indexes.append(max_score_idxs)
   
            
```

```python
# here the indexes are in tensor form
similar_prod_indexes[0:3]
```

```python
#To create numeric indexes
similar_prod_indexes = [x.item() for x in similar_prod_indexes]
```

```python
similar_prod_indexes[0:3]
```

```python
amazon_data.product_name[19995], flipkart_data.product_name[similar_prod_indexes[19995]]
```

# Matching data between amazon and flipkart

```python
# Extracting the details such as retail price and discounted price for both Flipkart and Amazon similar products
flipkart_prod_data = []
amazon_prod_data = []

for idx, prod_name in enumerate(amazon_data.product_name):
    flipkart_prod_data.append(flipkart_data.iloc[similar_prod_indexes[idx]])
    amazon_prod_data.append(amazon_data.iloc[idx])
```

```python
flipkart_prod_data = pd.DataFrame(flipkart_prod_data)
flipkart_prod_data.columns = ['Product name in Flipkart','Retail Price in Flipkart','Discounted Price in Flipkart']
amazon_prod_data = pd.DataFrame(amazon_prod_data)
amazon_prod_data.columns = ['Product name in Amazon','Retail Price in Amazon','Discounted Price in Amazon']
```

```python
flipkart_prod_data.head()
```

```python
flipkart_prod_data = flipkart_prod_data.reset_index(drop=True)
```

```python
amazon_prod_data.head()
```

```python
# concatenating data
final_prod_data = pd.concat([flipkart_prod_data,amazon_prod_data],axis=1)
final_prod_data.to_csv('final_result.csv')
```
